```python
import datetime as dt

def jul_create_instances(jul_path):
  dirs, _ = os.path.split(jul_path)
  output_file = os.path.join(dirs, "instance_output")

  f_data = open(jul_path, "r")
  user_dict = {}
  item_list = []
  first = True
  
  for line in f_data:
    if first:
      first = False
      continue
  
    line = line.strip()
    things = line.split(",")
  
    date_ts = str(int(dt.datetime(int(things[-3]),int(things[-2]),int(things[-1])).timestamp()))
    line1 = things[0] + "\t" + things[4]+ "\t" + date_ts + "\t" + things[6]
  
    if things[0] not in user_dict:
        user_dict[things[0]] = []
    user_dict[things[0]].append((line1, float(date_ts)))
    item_list.append(things[1])
  
  f_output = open(output_file, "w")
  for user_behavior in user_dict:
      sorted_user_behavior = sorted(user_dict[user_behavior], key=lambda x: x[1])
      for line, _ in sorted_user_behavior:
          user_things = line.split("\t")          
           
          f_output.write("1" + "\t" + line + "\n") 
    
  f_output.close()
  return output_file
```

```python
def encode_column(column):
    """ Encodes a pandas column with continuous IDs"""
    keys = column.unique()
    key_to_id = {key:idx for idx,key in enumerate(keys)}
    id_to_key = {idx:key for idx,key in enumerate(keys)}
    return key_to_id, id_to_key, np.array([key_to_id[x] for x in column]), len(keys)

def encode_df(df):
    """Encodes quantity data with continuous user and prod ids"""

    hist_ids, id_hists, df['hist_it'], num_hist = encode_column(df['hist_it'])
    item_ids, id_items, df['item_id'], num_items = encode_column(df['item_id'])
    user_ids, id_users, df['user_id'], num_users = encode_column(df['user_id'])
    return df, num_users, num_hist, num_items, user_ids, hist_ids, item_ids,  id_hists, id_users, id_items

```

```python
from scipy import sparse
def create_embeddings(n, K):
    """
    Creates a random numpy matrix of shape n, K with uniform values in (0, 11/K)
    n: number of items/users
    K: number of factors in the embedding 
    """
    return 11*np.random.random((n, K)) / K

def create_sparse_matrix(df, rows, cols, column_name="item_id"): # we treat quantity as rating of a product
    """ Returns a sparse utility matrix""" 
    return sparse.csc_matrix((df[column_name].values,(df['user_id'].values, df['hist_it'].values)),shape=(rows, cols))
```

```python
Y = create_sparse_matrix(dfp, num_users, num_hist)
Y.todense()
```

```python
def predict(df, emb_user, emb_prod):
    """ This function computes df["prediction"] without doing (U*V^T).
    
    Computes df["prediction"] by using elementwise multiplication of the corresponding embeddings and then 
    sum to get the prediction u_i*v_j. This avoids creating the dense matrix U*V^T.
    """
    df['prediction'] = np.round(np.sum(np.multiply(emb_prod[df['hist_it']],emb_user[df['user_id']]), axis=1))
    # print(np.round(np.sum(np.multiply(emb_prod[df['hist_it']],emb_user[df['user_id']]), axis=1)))
    return df
```

```python
lmbda = 0.0002
```

```python
from sklearn.metrics import log_loss

def cost(df, emb_user, emb_prod):
    """ Computes mean square error"""
    Y = create_sparse_matrix(df, emb_user.shape[0], emb_prod.shape[0])
    predicted = create_sparse_matrix(predict(df, emb_user, emb_prod), emb_user.shape[0], emb_prod.shape[0], 'prediction')
    return log_loss(Y.todense(), predicted.todense())#np.sum((Y-predicted).power(2))/df.shape[0] 

def gradient(df, emb_user, emb_prod):
    """ Computes the gradient for user and item embeddings"""
    Y = create_sparse_matrix(df, emb_user.shape[0], emb_prod.shape[0])
    predicted = create_sparse_matrix(predict(df, emb_user, emb_prod), emb_user.shape[0], emb_prod.shape[0], 'prediction')
    delta =(Y-predicted)
    grad_user = (-2/df.shape[0])*(delta*emb_prod) + 2*lmbda*emb_user
    grad_prod = (-2/df.shape[0])*(delta.T*emb_user) + 2*lmbda*emb_prod
    return grad_user, grad_prod

def gradient_descent(df, emb_user, emb_prod, iterations=3000, learning_rate=0.01, df_val=None):
    """ 
    Computes gradient descent with momentum (0.9) for given number of iterations.
    emb_user: the trained user embedding
    emb_prod: the trained prod embedding
    """
    Y = create_sparse_matrix(df, emb_user.shape[0], emb_prod.shape[0])
    beta = 0.9
    grad_user, grad_anime = gradient(df, emb_user, emb_prod)
    v_user = grad_user
    v_anime = grad_anime
    for i in range(iterations):
        grad_user, grad_anime = gradient(df, emb_user, emb_prod)
        v_user = beta*v_user + (1-beta)*grad_user
        v_anime = beta*v_anime + (1-beta)*grad_anime
        emb_user = emb_user - learning_rate*v_user
        emb_prod = emb_prod - learning_rate*v_anime
        if(not (i+1)%50):
            print("\niteration", i+1, ":")
            print("train mse:",  cost(df, emb_user, emb_prod))
            if df_val is not None:
                print("validation mse:",  cost(df_val, emb_user, emb_prod))
    return emb_user, emb_prod    
```

```python
emb_user = create_embeddings(num_users, 127)
emb_prod = create_embeddings(num_hist, 127)
emb_user, emb_prod = gradient_descent(dfp, emb_user, emb_prod, iterations=5000, learning_rate=0.1)
```

```python
def encode_new_data(valid_df, user_ids, prod_ids):
    """ Encodes valid_df with the same encoding as train_df.
    """
    df_val_chosen = valid_df['hist_it'].isin(prod_ids.keys()) & valid_df['user_id'].isin(user_ids.keys())
    valid_df = valid_df[df_val_chosen]
    valid_df['hist_it'] =  np.array([prod_ids[x] for x in valid_df['hist_it']])
    valid_df['user_id'] = np.array([user_ids[x] for x in valid_df['user_id']])
    return valid_df
```

```python
class LoadFrozedPredModel:
    def __init__(self, graph):
        self.pred = graph.get_tensor_by_name('import/pred:0') 
        self.items = graph.get_tensor_by_name('import/items:0') 
        self.cates = graph.get_tensor_by_name('import/cates:0') 
        self.item_history = graph.get_tensor_by_name('import/item_history:0') 
        self.item_cate_history = graph.get_tensor_by_name('import/item_cate_history:0') 
        self.mask = graph.get_tensor_by_name('import/mask:0')  
        self.time_from_first_action = graph.get_tensor_by_name('import/time_from_first_action:0') 
        self.time_to_now = graph.get_tensor_by_name('import/time_to_now:0') 
        self.layer_keeps = graph.get_tensor_by_name('import/layer_keeps:0') 
        self.is_training = graph.get_tensor_by_name('import/is_training:0')
```

```python
def infer_as_serving(model, infile, outfile, hparams, iterator, sess):
    preds = []
    
    for batch_data_input in iterator.load_data_from_file(infile, batch_num_ngs=0):
        if batch_data_input:
            feed_dict = {
                model.layer_keeps:np.ones(3, dtype=np.float32),
                model.is_training:False,
                model.items: batch_data_input[iterator.items],
                model.cates: batch_data_input[iterator.cates],
                model.item_history: batch_data_input[iterator.item_history],
                model.item_cate_history: batch_data_input[iterator.item_cate_history],
                model.mask: batch_data_input[iterator.mask],
                model.time_from_first_action: batch_data_input[iterator.time_from_first_action],
                model.time_to_now: batch_data_input[iterator.time_to_now]
            }
            step_pred = sess.run(model.pred, feed_dict=feed_dict)
            print(step_pred)
            preds.extend(np.reshape(step_pred, -1))
                
    with open(outfile, "w") as wt:
        for line in preds:
            wt.write('{0}\n'.format(line))
```
